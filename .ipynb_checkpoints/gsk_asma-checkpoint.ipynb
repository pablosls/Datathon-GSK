```python
cols = ['Valor_total', 'Valor_serviços_hospitalares', 
        'Val_serv_hosp_-_compl_federal', 'Val_serv_hosp_-_compl_gestor', 
       'Valor_serviços_profissionais','Val_serv_prof_-_compl_federal', 'Val_serv_prof_-_compl_gestor', 
       'Valor_médio_AIH', 'Valor_médio_intern', 'Taxa_mortalidade', 'Dias_permanência', 'Média_permanência', 'Óbitos'] 
for col in cols:  
    df[col] = df[col].replace('-','0')
    df[col] = pd.to_numeric(df[col])
```
